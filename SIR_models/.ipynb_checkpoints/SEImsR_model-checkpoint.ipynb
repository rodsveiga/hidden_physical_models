# The SEI(ms)R model

This set is our first shot:

$$  \frac{dS}{dt} = - \frac{\beta S}{N} ( \alpha_m I_m + \alpha_s I_s)        $$

$$  \frac{dE}{dt} =  \frac{\beta S}{N} ( \alpha_m I_m + \alpha_s I_s) - ( \kappa_m +  \kappa_S ) E        $$

$$  \frac{d I_m}{dt} = \kappa_m E  - \alpha_m p_m I_m - \tilde{\alpha}_m (1 - p_m) I_m  $$

$$  \frac{d I_s}{dt} = \kappa_s E  - \tilde{\alpha}_m (1 - p_m) I_s - \alpha_s p_s I_s - \tilde{\alpha}_s ( 1 - p_s) I_s  $$

$$  \frac{d R}{dt} = \alpha_m p_m I_m + \alpha_s p_s I_s  $$

$$  \frac{d D}{dt} = \tilde{\alpha}_s (1 - p_s) I_s  $$

with $  N(t) = S(t) + E(t) + I_{m}(t) +  I_{s}(t) + R(t) + D(t)$.

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# Total population
N = 10000

# beta: contact rate
# alpha_m: average time mild infected patients are infectious, in 1/days
# alpha_s: average time severe infected patients are infectious, in 1/days
# kappa_m: average incubation time for mild infected patients, in 1/days
# kappa_s: average incubation time for severe infected patients, in 1/days
beta, alpha_m, alpha_s, kappa_m, kappa_s, p_m, p_s = 3.0, 1./14, 1./21, 1./7, 1./7, 0.9, 0.4

# I_m0: Initial number of infected with mild symptoms
# I_s0: Initial number of infected with severe symptoms
# E0: Initial number of exposed individuals
# R0: Initial number of recovered fro the epidemic
# D0: Initial number of deaths from the epidemic
I_m0, I_s0, E0, R0, D0 = 10, 1, 0, 0, 0

# Everyone else, S0, is susceptible to infection initially.
S0 = N - E0 - R0 - D0 - I_m0 - I_s0

# Grid of time poins, in days
t = np.linspace(0, 250, 250)
```

```python
def deriv(y, t, N, params):
    
    S, E, I_m, I_s, R, D = y
    beta, alpha_m, alpha_s, kappa_m, kappa_s, p_m, p_s = params
   
    
    dSdt = - beta*S*( alpha_m * I_m + alpha_s * I_s )/N
    
    dEdt = beta*S*( alpha_m * I_m + alpha_s * I_s )/N - (kappa_m + kappa_s)*E
    
    dImdt = kappa_m*E - alpha_m*p_m*I_m - alpha_m*(1. - p_m)*I_m
    
    dIsdt = kappa_s*E + alpha_m*(1.-p_m)*I_s - alpha_s*p_s*I_s - alpha_s*(1-p_s)*I_s
    
    dRdt = alpha_m*p_m*I_m + alpha_s*p_s*I_s
    
    dDdt = alpha_s*(1. - p_s)*I_s
    
    return dSdt, dEdt, dImdt, dIsdt, dRdt, dDdt
```

```python
# Initial conditions vector
y0 = S0, E0, I_m0, I_s0, R0, D0
```

```python
# Parameters
params = beta, alpha_m, alpha_s, kappa_m, kappa_s, p_m, p_s
```

```python
# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, params))
S, E, I_m, I_s, R, D = ret.T
```

```python
plt.plot(t, S/N, color= 'C0', label='Susceptible')
plt.plot(t, E/N, color= 'C1', label='Exposed')
plt.plot(t, I_m/N, color= 'C2', label='Infected mild')
plt.plot(t, I_s/N, color= 'C3', label='Infected severe')
plt.plot(t, R/N, color= 'C4', label='Recovered')
plt.plot(t, D/N, color= 'C5', label='Deaths')
plt.xlabel('Time (days)', fontsize= 12)
plt.ylabel('Population fraction', fontsize= 12)
#plt.xlim([-5, 200])
plt.grid(linestyle=':')
plt.legend(loc= 'lower left', fontsize= 8)
plt.show()
```
